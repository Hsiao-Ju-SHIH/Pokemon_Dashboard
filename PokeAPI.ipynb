```python
import numpy as np
import pandas as pd
import pokebase as pb
```

```python
dfs = []
for i in range(1, 1026):
    pokemon = pb.pokemon(i)
    name = pokemon.name
    height = pokemon.height
    weight = pokemon.weight
    stats = []
    stats.append(name)
    stats.append(height)
    stats.append(weight)
    for stat in pokemon.stats:
        stats.append(stat.base_stat)
    for type in pokemon.types:
        name = type.type.name
        stats.append(name)
    for ability in pokemon.abilities:
        name_ab = ability.ability.name
        stats.append(name_ab)
    stats = pd.Series(stats)
    dfs.append(stats)
data = pd.concat(dfs, axis=1).T
data.columns = ['Name', 'Height', 'Weight', 'HP', 'Attack', 'Defense', 'Sp. Attack', 'Sp. Defense', 'Speed', 'Type 1', 'Type 2', 'Ability 1', 'Ability 2', 'Ability 3']
data
```

```python
dfs = []
for i in range(1, 10):
    color = pb.pokemon_color(i)
    name = color.name
    mons = []
    for mon in color.pokemon_species:
        mons.append(mon.name)
    df = pd.DataFrame(mons, columns=[name])
    dfs.append(df)
data_color = pd.concat(dfs, axis=1)
melt_data_color = data_color.melt(var_name='color', value_name='pokemon').dropna()
melt_data_color
```

```python
dfs = []
for i in range(1, 15):
    shape = pb.pokemon_shape(i)
    name = shape.name
    mons = []
    for mon in shape.pokemon_species:
        mons.append(mon.name)
    df = pd.DataFrame(mons, columns=[name])
    dfs.append(df)
data_shape = pd.concat(dfs, axis=1)
melt_data_shape = data_shape.melt(var_name='shape', value_name='pokemon').dropna()
melt_data_shape
```

```python
df_fin = (data.merge(melt_data_color, how='left', left_on='Name', right_on='pokemon')).merge(melt_data_shape, how='left', left_on='Name', right_on='pokemon')
df_fin['Height (m)'] = df_fin['Height'] / 10
df_fin['Weight (kg)'] = df_fin['Weight'] / 10
df_fin = df_fin.loc[:, ~df_fin.columns.isin(['pokemon_x', 'pokemon_y', 'Height', 'Weight'])]
df_fin.to_csv('pokemon.csv', index=False)
```
