```python
import pandas as pd
```

```python
df = pd.read_csv('pokemon.csv')
df_spe = pd.read_csv('pokemon_species.csv')
```

```python
df_merge = df.merge(df_spe.loc[:, df_spe.columns != 'Name'], left_on='Index', right_on='ID')
df_merge = df_merge.loc[:, df_merge.columns != 'Index']
df_merge
```

```python
df_merge.to_csv('pokemon_updated.csv', index=False)
```
